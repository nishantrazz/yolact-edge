```python
from ultralytics import YOLO
import cv2
import time

# Load the exported TensorRT model with instance segmentation
model = YOLO("canopy11s_rect.pt", task="segment")  # Enable segmentation

# Open the video file
video_path = "Dataset_Canopy/DJI_0008.MOV"  # Replace with your video file path
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_width = 640 # int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = 360 # int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Initialize time and frame count for FPS calculation
prev_time = time.time()
frame_count = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Break the loop if video ends

    # Resize the frame to 640x640
    frame_resized = cv2.resize(frame, (frame_width, frame_height))

    # Run inference on the resized frame
    results = model(frame_resized)

    # Draw the detections on the frame
    for result in results:
        annotated_frame = result.plot()

    # Calculate FPS
    frame_count += 1
    current_time = time.time()
    elapsed_time = current_time - prev_time
    if elapsed_time > 0:
        fps = frame_count / elapsed_time
    else:
        fps = 0

    # Display FPS on the frame
    cv2.putText(annotated_frame, f"FPS: {fps:.2f}", (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 
                1, (0, 255, 0), 2, cv2.LINE_AA)

    # Display the frame
    cv2.imshow("YOLO Instance Segmentation", annotated_frame)
    
    if cv2.waitKey(1) & 0xFF == ord("q"):  # Press 'q' to exit early
        break

# Release everything
cap.release()
cv2.destroyAllWindows()
```
